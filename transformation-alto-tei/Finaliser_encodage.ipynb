Finaliser l'encodage d'une lettre
==========================

Ce *notebook* énumère les tâches à accomplir par intervention manuelle dans les fichiers TEI pour finaliser l'encodage d'une lettre.

Une fois l'encodage automatique effectué, il est nécessaire de déplacer les fichiers du dossier ./data/ vers un nouveau dossier, afin d'empêcher tout écrasement des tâches à suivre.

# Structuration en paragraphes

Les éléments `p` doivent être divisés selon la mise en page de la lettre : un `p` pour chaque paragraphe.


# Corrections interlinéaires

![correction](./demo/CdS02_Konv002-03_0072-detail-corr.jpg)

Les corrections interlinéaires sont indiquées par un commentaire et dans un élément `choice` (les éléments `lb` ont été supprimés des exemples pour les rendre plus lisibles) :

```xml
l'éxaltation d'une âme [........] tendre et
<!--Correction interlinéaire-->
<choice>
  <sic/>
  <corr>
    vraiment
  </corr>
</choice>
passionnée
```

L'élément `choice` doit être déplacé et la partie corrigée doit venir compléter le `sic` ; en l'espèce, la partie corrigée est illisible. Une fois la correction manuelle effectuée, on obtient :

```xml
l'éxaltation d'une âme 
<choice>
  <sic>
    <unclear reason="illegible"/>
  </sic>
  <corr>
    vraiment
  </corr>
</choice> tendre et passionnée
```


# Notes, éléments marginaux

Les lignes appartenant à des MarginTextZone ou NumberginZone sont ajoutées à l'encodage à la fin de la page.

## Lorsque la lettre se poursuit sur la page suivante

![notes et numérotation en bas de page](./demo/CdS02_Konv002-02_0027-detail-basPage.jpg)

Dans ce cas l'encodage automatique ferme l'élément `p` ou `lg` en cours, insère les éléments de type note ou numérotation et ouvre un nouvel élément pour la page suivante :

```xml
<p>(…) Nous en gardons 3 depuis 3 jours, mais très aimable. Un colonel du</p>
<note type="MarginTextZone">(Corresp : part.)
    <!--Vérifier que la note corresponde à la lettre et mettre à jour le type-->
</note>
<fw type="NumberingZone">
    19
    <!--Vérifier que le numéro corresponde à la lettre et mettre à jour le type : pageNum si numéro de page, letterNum si numéro de lettre-->
</fw>
<pb corresp="#CdS02_Konv002-02_0028"/>
<p>meilleur ton, homme de société et spirituelle. Un beau major de (…)</p>
```

A noter que le type de région d'écriture est rappelé par l'`@type` des éléments `note` et `fw` qui contiennent respectivement les notes marginales et les numérotations.

Des commentaires sont automatiquement intégrés à l'encodage pour guider le travail. La liste des opérations à faire est la suivante :

- `fw` : pour l'exemple présent `@type="letterNum"`
- `note` : pour l'exemple présent `@type="short"`
- `p` : réunir tous les éléments dans un même `p` si le saut de page ne correspond pas à un saut de paragraphe.

L'encodage final ressemble à ce ceci :

```xml
<p>
  (…)
  Nous en gardons 3 depuis 3 jours, mais très aimable. Un colonel du
  <note type="short">(Corresp : part.)</note>
  <fw type="letterNum">19</fw>
  <pb corresp="#CdS02_Konv002-02_0028"/>
  meilleur ton, homme de société et spirituelle. Un beau major de
  (…)
</p>
```

# Corriger les fautes manifestes

![faute](./demo/CdS02_Konv002-02_0027-detail-faute.jpg)

Dans cet exemple on ne tiendra pas compte de la correction interlinéaire pour la clarté de l'explication. L'encodage automatique donne :

```xml
Je vais que vous avez aussi des personnages de différentes espèces.
```

On utilise l'élément TEI `app` pour corriger le texte :

```xml
Je 
<app>
    <lem>vois</lem>
    <rdg>vais</rdg>
</app> 
que vous avez aussi des personnages de différentes espèces.
```

# Mention de date mal placée

Il peut arriver que la ligne de type `dateline` soit mal numérotée et qu'on ne l'ait pas remarqué. Ce cas de figure est indiqué dans l'encodage de la façon suivante :

```xml
<p>
    <lb corresp="#CdS02_Konv002-03_0072_z2_l4"/>
    Madame, j'ai reçu le joli cadeau et l'aimable billet que vous avez
    <!--Mention de date mal placée dans l'ordre des lignes-->
    <lb corresp="#CdS02_Konv002-03_0072_z2_l5"/>
    Paris, ce 21 Juin, 1824.
    <lb corresp="#CdS02_Konv002-03_0072_z2_l6"/>
    bien voulu m'adresser le 2 de ce mois. Je vous prie d'en recevoir mes
    <lb corresp="#CdS02_Konv002-03_0072_z2_l7"/>
    très humbles remercimens.
</p>
```
